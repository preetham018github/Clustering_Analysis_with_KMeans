# KMeans Clustering

## Importing the relevant libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Set the styles to Seaborn
sns.set()
# Importing the KMeans module so we can perform k-means clustering with sklearn
from sklearn.cluster import KMeans
```

## Loading the data

```python
# Loading the data
data = pd.read_csv ('Clustering_Dataset.csv')
```

```python
# Checking what's inside
data
```

## Ploting the data

Creating a preliminary plot to see if we can spot something

```python
# We are creating a scatter plot of the two variables
plt.scatter(data['Satisfaction'],data['Loyalty'])
# Naming your axes 
plt.xlabel('Satisfaction')
plt.ylabel('Loyalty')
```

## Selecting the features

```python
# Selecting both features by creating a copy of the data variable
x = data.copy()
```

## Clustering

```python
# Creating an object (which we would call kmeans)
# The number in the brackets is K, or the number of clusters we are aiming for
kmeans = KMeans(2)
# Fiting the data
kmeans.fit(x)
```

## Clustering results

```python
# Creating a copy of the input data
clusters = x.copy()
# Taking note of the predicted clusters 
clusters['cluster_pred']=kmeans.fit_predict(x)
```

```python
# Ploting the data using the longitude and the latitude
# c (color) is an argument which could be coded with a variable 
# The variable in this case has values 0,1, indicating to plt.scatter, that there are two colors (0,1)
# All points in cluster 0 will be the same colour, all points in cluster 1 - another one, etc.
# cmap is the color map.
plt.scatter(clusters['Satisfaction'],clusters['Loyalty'],c=clusters['cluster_pred'],cmap='rainbow')
plt.xlabel('Satisfaction')
plt.ylabel('Loyalty')
```

## Standardize the variables

Let's standardize and check the new result

```python
# Import a library which can do that easily
from sklearn import preprocessing
# Scale the inputs
# preprocessing.scale scales each variable (column in x) with respect to itself
# The new result is an array
x_scaled = preprocessing.scale(x)
x_scaled
```

## Taking advantage of the Elbow method

```python
# Creating an empty list
wcss =[]

# Creating all possible cluster solutions with a loop
# We have chosen to get solutions from 1 to 9 clusters
for i in range(1,10):
    # Clsuter solution with i clusters
    kmeans = KMeans(i)
    # Fiting the STANDARDIZED data
    kmeans.fit(x_scaled)
    # Appending the WCSS for the iteration
    wcss.append(kmeans.inertia_)
    
# Checking the result
wcss
```

```python
# Ploting the number of clusters vs WCSS
plt.plot(range(1,10),wcss)
# Naming your axes
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
```

## Exploring clustering solutions and select the number of clusters

```python
kmeans_new = KMeans(4)
# Fiting the data
kmeans_new.fit(x_scaled)
# Creating a new data frame with the predicted clusters
clusters_new = x.copy()
clusters_new['cluster_pred'] = kmeans_new.fit_predict(x_scaled)
```

```python
# Checking if everything seems right
clusters_new
```

```python
# Plot
plt.scatter(clusters_new['Satisfaction'],clusters_new['Loyalty'],c=clusters_new['cluster_pred'],cmap='rainbow')
plt.xlabel('Satisfaction')
plt.ylabel('Loyalty')
```
